```python
import numpy as np 
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

```python
data = [['milk','bread','rice','book'],
        ['bread', 'jam', 'book', 'pen'], 
        ['jam','milk', 'bread','rice','aggs'],
        ['rice','eggs','pen','books'],
        ['eggs','pen','milk','bread','jam'],
        ['eggs', 'rice' ,'bread', 'jam']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te
```

```python
te_array = te.fit_transform(data)
te_array
```

```python
df = pd.DataFrame(te_array,columns = te.columns_).astype(int)
df # this output is called transectional encoder
```

```python
from mlxtend.frequent_patterns import apriori
itemset = apriori (df,min_support = 0.3, use_colnames = True)
itemset
```

```python
from mlxtend.frequent_patterns import association_rules
res = association_rules(itemset,metric = 'confidence',min_threshold=0.7)
res
```

```python
result = res[['antecedents','consequents','support','confidence','lift']]
result
```
